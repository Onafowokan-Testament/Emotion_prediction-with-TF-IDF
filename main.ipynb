```python
train_data['target'] = train_data['emotions'].map({
    'joy': 1,'sadness':2, 'anger':3,'fear':4, 'love':5, 'surprise':6
})
```

```python
test_data['target'] = test_data['emotions'].map({
    'joy': 1,'sadness':2, 'anger':3,'fear':4, 'love':5, 'surprise':6
})
```

```python
train_data
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.naive_bayes import MultinomialNB
```

//Let us firstly use the random forest as it is good for classification.

```python
rf_pipeline = Pipeline(
[
    ('vectorizer',CountVectorizer(ngram_range=(3,3))),
    ('rfc', RandomForestClassifier()),

]
)

model_rf = rf_pipeline.fit(train_data['comment'], train_data['target'])

y_pred_rf = model_rf.predict(test_data['comment'])

report_rf = classification_report(test_data['comment'], test_data['target'])

report_rf
```

```python
mnb_pipeline = Pipeline(
[
    ('vectorizer',CountVectorizer(ngram_range=(1,2))),
    ('mnb', MultinomialNB()),

]
)

model_mnb = rf_pipeline.fit(train_data['comment'], train_data['target'])

y_pred_mnb = model_rf.predict(test_data['comment'])

report_mnb = classification_report(test_data['comment'], test_data['target'])

report_mnb
```

```python

clf = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),        
     ('Random Forest', RandomForestClassifier())         
])


clf.fit(train_data['comment'], train_data['target'])


y_pred = clf.predict(test_data['comment'])



print(classification_report(test_data['comment'], test_data['target']))
```
